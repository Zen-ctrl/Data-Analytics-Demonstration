##**Q4 The dataset Real GDP.csv contains the GDP of each US State for all years starting from 1997 until 2020. The data is at State level, i.e., each observation corresponds to a unique State.**

The dataset **Surplus.csv** contains the surplus of each US State for all years starting from 1997 until 2020. The data is at year level, i.e., each observation corresponds to a unique year.

The dataset **Compensation.csv** contains ‘Compensation’ and ‘Chain-type quantity indexes for real GDP’ for each US State and year starting from 1997 to 2020. The dataset is at Year-State-Description level, where ‘Description’ refers to either ‘Compensation’ or ‘Chain-type quantity indexes for real GDP’


---



###**(a) Combine all these datasets to obtain a dataset at State-Year level, that contains the GDP, surplus, ‘Compensation’, and ‘Chain-type quantity indexes for real GDP’ for each US State and all years starting from 1997 until 2020. Note that each observation must contain the name of the US State, year, and the four values (GDP, surplus, compensation, and Chain-type quantity indexes for real GDP)**

```python
R_GDP = pd.read_csv('Real GDP.csv')

R_GDP2 = pd.melt(R_GDP, id_vars=['Description','GeoName'], var_name='Year', value_name='GDP')
R_GDP2 = R_GDP2.drop('Description', axis=1)
R_GDP2.Year = R_GDP2.Year.astype(np.int64)
R_GDP2.head()
```

```python
Surp = pd.read_csv('Surplus.csv')
Surp2 = pd.melt(Surp, id_vars=['Year'], var_name='State', value_name='Surplus')
Surp2 = Surp2.rename(columns={'State': 'GeoName'}, )
Surp2.head()
```

```python
Comp = pd.read_csv('Compensation.csv')
Comp2 = Comp[Comp['Description'] == "Chain-type quantity indexes for real GDP"]
Comp2
```

### **Merge the datasets**

```python
newpd = pd.merge(R_GDP2, Surp2, how='inner', on=['Year', 'GeoName'])
Q4 = pd.merge(newpd, Comp2, how='inner', on=['Year', 'GeoName'])
Q4
```

###**(c) The mean GDP (over all States) seems to have decreased in 2020 as compared to 2019 (you know why!). How many States observed a decrease in GDP in 2020 (as compared to 2019)? For which States did the GDP increase in 2020 (as compared to 2019)**

```python
df19 = Q4[Q4.Year == 2019].drop(['Surplus', 'value', 'Description', 'Year'], axis=1)
df20 = Q4[Q4.Year == 2020].drop(['Surplus', 'value', 'Description', 'Year'], axis=1)

df = pd.merge(df19, df20, how='inner', on='GeoName', suffixes=('_19', '_20'))
df['isIncreased'] = df['GDP_19'] < df['GDP_20']

print(df.isIncreased.value_counts())
print()
print(df[df.isIncreased == True].GeoName)
```

There are two states, South Dakota and Utah, which increased their GDP in 2020.

All the other states decreased their GDP.
